```python
# ============================================================
# PROTEIN INFORMATION SIGNATURES (HONEST VERSION)
# 20 real proteomes from UniProt (10 categories × 2 each)
# Real proteomes from UniProt + controls + paper-ready figures
# No clamping, no forced inequalities, no invented data.
# ============================================================

!pip -q install biopython pandas numpy requests tqdm

import os, io, math, random, gzip, requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from Bio import SeqIO

# ----------------------------
# CONFIG
# ----------------------------
OUTDIR = "/content/protein_info_honest_20proteomes"
os.makedirs(OUTDIR, exist_ok=True)

# 10 categories × 2 proteomes each = 20 proteomes
PROTEOMES = [
    # Mammals
    {"category": "Mammals", "name": "Homo sapiens (Human)", "id": "UP000005640"},
    {"category": "Mammals", "name": "Mus musculus (Mouse)", "id": "UP000000589"},

    # Birds
    {"category": "Birds", "name": "Gallus gallus (Chicken)", "id": "UP000000539"},
    {"category": "Birds", "name": "Taeniopygia guttata (Zebra finch)", "id": "UP000007754"},

    # Fish
    {"category": "Fish", "name": "Danio rerio (Zebrafish)", "id": "UP000000437"},
    {"category": "Fish", "name": "Takifugu rubripes (Pufferfish)", "id": "UP000005226"},

    # Amphibians
    {"category": "Amphibians", "name": "Xenopus tropicalis", "id": "UP000008143"},
    {"category": "Amphibians", "name": "Xenopus laevis", "id": "UP000186698"},

    # Insects
    {"category": "Insects", "name": "Drosophila melanogaster", "id": "UP000000803"},
    {"category": "Insects", "name": "Anopheles gambiae", "id": "UP000007062"},

    # Plants
    {"category": "Plants", "name": "Arabidopsis thaliana", "id": "UP000006548"},
    {"category": "Plants", "name": "Oryza sativa (Rice)", "id": "UP000059680"},

    # Bacteria
    {"category": "Bacteria", "name": "Escherichia coli K-12", "id": "UP000000625"},
    {"category": "Bacteria", "name": "Bacillus subtilis 168", "id": "UP000001570"},

    # Archaea
    {"category": "Archaea", "name": "Haloferax volcanii", "id": "UP000008243"},
    {"category": "Archaea", "name": "Saccharolobus solfataricus P2", "id": "UP000001974"},

    # Filamentous fungi
    {"category": "Fungi", "name": "Aspergillus nidulans", "id": "UP000000560"},
    {"category": "Fungi", "name": "Neurospora crassa", "id": "UP000001805"},

    # Yeasts
    {"category": "Yeasts", "name": "Saccharomyces cerevisiae", "id": "UP000002311"},
    {"category": "Yeasts", "name": "Schizosaccharomyces pombe", "id": "UP000002485"},
]

MIN_LEN = 80

# Cap sequences per proteome to keep Colab fast (still real data)
# Big eukaryotes are capped lower to avoid RAM/time explosions.
DEFAULT_CAP = 2500
MAX_SEQS_BY_PROTEOME = {
    # Mammals
    "UP000005640": 3500,  # human (very large)
    "UP000000589": 3500,  # mouse

    # Birds
    "UP000000539": 3000,  # chicken
    "UP000007754": 3000,  # zebra finch

    # Fish
    "UP000000437": 3000,  # zebrafish
    "UP000005226": 3000,  # fugu

    # Amphibians
    "UP000008143": 3000,  # x. tropicalis
    "UP000186698": 3000,  # x. laevis

    # Insects
    "UP000000803": 2500,  # drosophila
    "UP000007062": 2500,  # anopheles

    # Plants
    "UP000006548": 3000,  # arabidopsis
    "UP000059680": 3000,  # rice

    # Bacteria
    "UP000000625": 2500,  # ecoli
    "UP000001570": 2500,  # b subtilis

    # Archaea
    "UP000008243": 2500,  # haloferax
    "UP000001974": 2500,  # saccharolobus

    # Fungi
    "UP000000560": 2500,  # aspergillus
    "UP000001805": 2500,  # neurospora

    # Yeasts
    "UP000002311": 3000,  # s cerevisiae
    "UP000002485": 3000,  # s pombe
}

# Bootstrap
N_BOOT = 150
BOOT_SAMPLE_SIZE = 200
SEED = 123
np.random.seed(SEED)
random.seed(SEED)

# Info profile
MAX_D = 25
ID_SEQ_CAP = 120

# Compressibility
COMP_SEQ_CAP = 200
COMP_MIN_TOTAL_AA = 50_000

LN2 = np.log(2.0)
AA = "ACDEFGHIKLMNPQRSTVWY"
AA_SET = set(AA)
AA_TO_I = {a:i for i,a in enumerate(AA)}
AA_ARRAY = np.array(list(AA))
HMAX_BITS = math.log2(len(AA))

print("="*70)
print("PROTEIN INFORMATION SIGNATURES (HONEST) - 20 PROTEOMES")
print("="*70)
print("OUTDIR:", OUTDIR)
print(f"N_BOOT={N_BOOT} | BOOT_SAMPLE_SIZE={BOOT_SAMPLE_SIZE} | MAX_D={MAX_D}")
print(f"Id subsample cap={ID_SEQ_CAP} | Compress cap={COMP_SEQ_CAP} | Hmax={HMAX_BITS:.4f} bits/site")
print("="*70)

# ----------------------------
# DOWNLOAD + CLEAN
# ----------------------------
def download_proteome_fasta(proteome_id: str) -> str:
    url = "https://rest.uniprot.org/uniprotkb/stream"
    params = {"query": f"(proteome:{proteome_id})", "format": "fasta"}
    r = requests.get(url, params=params, timeout=240)
    r.raise_for_status()
    return r.text

def load_clean_seqs(fasta_text: str, max_seqs=None, min_len=80):
    seqs = []
    with io.StringIO(fasta_text) as handle:
        for rec in SeqIO.parse(handle, "fasta"):
            s = str(rec.seq).upper()
            if len(s) < min_len:
                continue
            if set(s).issubset(AA_SET):
                seqs.append(s)
                if max_seqs is not None and len(seqs) >= max_seqs:
                    break
    return seqs

proteome_data = {}
for p in PROTEOMES:
    pid, name, cat = p["id"], p["name"], p["category"]
    cap = MAX_SEQS_BY_PROTEOME.get(pid, DEFAULT_CAP)
    print(f"Downloading [{cat}] {name} ({pid}) ... cap={cap}")
    fasta_text = download_proteome_fasta(pid)
    seqs = load_clean_seqs(fasta_text, max_seqs=cap, min_len=MIN_LEN)
    if len(seqs) < 200:
        raise RuntimeError(f"Too few sequences kept for {name} ({pid}): {len(seqs)}. Lower MIN_LEN or raise cap.")
    lens = np.array([len(s) for s in seqs], dtype=int)
    proteome_data[pid] = {"category": cat, "name": name, "id": pid, "seqs": seqs, "lengths": lens}
    print(f"  kept {len(seqs)} sequences | median L={int(np.median(lens))}")

# ----------------------------
# CORE COUNTS
# ----------------------------
def counts1(seqs):
    c = np.zeros(20, dtype=np.int64)
    total = 0
    for s in seqs:
        idx = np.fromiter((AA_TO_I[ch] for ch in s), dtype=np.int16)
        np.add.at(c, idx, 1)
        total += len(s)
    return c, total

def C2_adjacent(seqs):
    C = np.zeros((20,20), dtype=np.int64)
    total = 0
    for s in seqs:
        if len(s) < 2:
            continue
        idx = np.fromiter((AA_TO_I[ch] for ch in s), dtype=np.int16)
        a = idx[:-1]
        b = idx[1:]
        np.add.at(C, (a,b), 1)
        total += len(idx)-1
    return C, total

def Cd_distance(seqs, max_d):
    Cd = np.zeros((max_d,20,20), dtype=np.int64)
    for s in seqs:
        idx = np.fromiter((AA_TO_I[ch] for ch in s), dtype=np.int16)
        L = len(idx)
        for d in range(1, max_d+1):
            if L <= d:
                break
            a = idx[:-d]
            b = idx[d:]
            np.add.at(Cd[d-1], (a,b), 1)
    return Cd

def H_from_p(p):
    p = np.asarray(p, dtype=float)
    p = p[p > 0]
    return float(-np.sum(p * np.log(p)) / LN2)

def MI_from_joint_counts(Cxy, p1=None):
    Cxy = Cxy.astype(np.float64)
    tot = Cxy.sum()
    if tot <= 0:
        return 0.0
    pxy = Cxy / tot
    if p1 is None:
        px = pxy.sum(axis=1, keepdims=True)
        py = pxy.sum(axis=0, keepdims=True)
    else:
        p1 = np.asarray(p1, dtype=np.float64)
        px = p1.reshape(-1,1)
        py = p1.reshape(1,-1)
    denom = px * py
    mask = (pxy > 0) & (denom > 0)
    mi_nats = np.sum(pxy[mask] * np.log(pxy[mask] / denom[mask]))
    return float(mi_nats / LN2)

def H1_I1(seqs):
    c1, total = counts1(seqs)
    p1 = c1 / total
    H1_bits = H_from_p(p1)
    C2, _ = C2_adjacent(seqs)
    I1_bits = MI_from_joint_counts(C2, p1=p1)
    return p1, H1_bits, I1_bits, C2

def Id_profile(seqs, max_d):
    c1, total = counts1(seqs)
    p1 = c1 / total
    Cd = Cd_distance(seqs, max_d)
    Id = np.zeros(max_d, dtype=float)
    for d in range(max_d):
        Id[d] = MI_from_joint_counts(Cd[d], p1=p1)
    return Id

# ----------------------------
# CONTROLS
# ----------------------------
def sample_iid(p1, lengths):
    out = []
    for L in lengths:
        draws = np.random.choice(AA_ARRAY, size=int(L), p=p1)
        out.append("".join(draws))
    return out

def transition_matrix(C2, alpha=1e-12):
    C = C2.astype(np.float64) + alpha
    C = C / C.sum(axis=1, keepdims=True)
    return C

def sample_markov1(p_init, T, lengths):
    out = []
    for L in lengths:
        L = int(L)
        first = np.random.choice(AA_ARRAY, p=p_init)
        seq = [first]
        cur = AA_TO_I[first]
        for _ in range(L-1):
            nxt = np.random.choice(AA_ARRAY, p=T[cur])
            seq.append(nxt)
            cur = AA_TO_I[nxt]
        out.append("".join(seq))
    return out

# ----------------------------
# COMPRESSIBILITY
# ----------------------------
GZIP_OVERHEAD_BITS = len(gzip.compress(b"")) * 8

def concat_for_compression(seqs, cap=200, min_total_aa=50_000):
    if not seqs:
        return ""
    pick = seqs[:] if len(seqs) <= cap else random.sample(seqs, cap)
    random.shuffle(pick)
    parts = []
    total = 0
    for s in pick:
        parts.append(s)
        total += len(s)
        if total >= min_total_aa:
            break
    return "|".join(parts)

def gzip_bits_per_symbol(long_string: str):
    if not long_string:
        return np.nan, 0
    b = long_string.encode("ascii")
    comp = gzip.compress(b, compresslevel=6)
    comp_bits = len(comp) * 8
    L = len(long_string)
    eff_bits = comp_bits - GZIP_OVERHEAD_BITS
    return eff_bits / L, L

# ----------------------------
# BOOTSTRAP PIPELINE
# ----------------------------
rows = []
id_store = []

print("\nRunning bootstrap...")

for pid, info in proteome_data.items():
    name = info["name"]
    cat = info["category"]
    seqs_all = info["seqs"]
    n = len(seqs_all)

    print(f"\n[{cat}] {name} ({pid}): n={n} | reps={N_BOOT} | sample={BOOT_SAMPLE_SIZE}")
    for b in tqdm(range(N_BOOT), desc=name[:18]):
        idx = np.random.randint(0, n, size=BOOT_SAMPLE_SIZE)
        boot_seqs = [seqs_all[i] for i in idx]
        boot_lengths = np.array([len(s) for s in boot_seqs], dtype=int)

        # REAL
        p1_r, H1_r, I1_r, C2_r = H1_I1(boot_seqs)
        Hrate_r = max(H1_r - I1_r, 0.0)

        s_long_r = concat_for_compression(boot_seqs, cap=COMP_SEQ_CAP, min_total_aa=COMP_MIN_TOTAL_AA)
        Cgz_r, Lr = gzip_bits_per_symbol(s_long_r)

        # IID control
        iid_seqs = sample_iid(p1_r, boot_lengths)
        p1_i, H1_i, I1_i, C2_i = H1_I1(iid_seqs)
        Hrate_i = max(H1_i - I1_i, 0.0)

        s_long_i = concat_for_compression(iid_seqs, cap=COMP_SEQ_CAP, min_total_aa=COMP_MIN_TOTAL_AA)
        Cgz_i, Li = gzip_bits_per_symbol(s_long_i)

        # Markov-1 control
        T = transition_matrix(C2_r)
        mk_seqs = sample_markov1(p1_r, T, boot_lengths)
        p1_m, H1_m, I1_m, C2_m = H1_I1(mk_seqs)
        Hrate_m = max(H1_m - I1_m, 0.0)
        Cgz_m = np.nan  # keep optional/off

        # Id profiles (declared subsample)
        boot_seqs_id = boot_seqs if len(boot_seqs) <= ID_SEQ_CAP else random.sample(boot_seqs, ID_SEQ_CAP)
        iid_seqs_id  = iid_seqs  if len(iid_seqs)  <= ID_SEQ_CAP else random.sample(iid_seqs,  ID_SEQ_CAP)
        mk_seqs_id   = mk_seqs   if len(mk_seqs)   <= ID_SEQ_CAP else random.sample(mk_seqs,   ID_SEQ_CAP)

        Id_r = Id_profile(boot_seqs_id, MAX_D)
        Id_i = Id_profile(iid_seqs_id,  MAX_D)
        Id_m = Id_profile(mk_seqs_id,   MAX_D)

        for d in range(MAX_D):
            id_store.append({"proteome_id": pid, "proteome_name": name, "category":cat, "ensemble":"Real", "rep":b, "d":d+1, "Id":Id_r[d]})
            id_store.append({"proteome_id": pid, "proteome_name": name, "category":cat, "ensemble":"IID",  "rep":b, "d":d+1, "Id":Id_i[d]})
            id_store.append({"proteome_id": pid, "proteome_name": name, "category":cat, "ensemble":"Markov-1","rep":b, "d":d+1, "Id":Id_m[d]})

        rows.append({"proteome_id":pid,"proteome_name":name,"category":cat,"rep":b,"ensemble":"Real",
                     "H1":H1_r,"I1":I1_r,"Hrate":Hrate_r,"Cgz":Cgz_r,"Cgz_L":Lr})
        rows.append({"proteome_id":pid,"proteome_name":name,"category":cat,"rep":b,"ensemble":"IID",
                     "H1":H1_i,"I1":I1_i,"Hrate":Hrate_i,"Cgz":Cgz_i,"Cgz_L":Li})
        rows.append({"proteome_id":pid,"proteome_name":name,"category":cat,"rep":b,"ensemble":"Markov-1",
                     "H1":H1_m,"I1":I1_m,"Hrate":Hrate_m,"Cgz":Cgz_m,"Cgz_L":np.nan})

df_points = pd.DataFrame(rows)
df_id = pd.DataFrame(id_store)

# Save CSVs
points_csv = os.path.join(OUTDIR, "bootstrap_points.csv")
df_points.to_csv(points_csv, index=False)

id_csv = os.path.join(OUTDIR, "Id_profile_mean_std.csv")
df_id_sum = (df_id.groupby(["proteome_id","proteome_name","category","ensemble","d"], as_index=False)
             .agg(Id_mean=("Id","mean"), Id_std=("Id","std")))
df_id_sum.to_csv(id_csv, index=False)

summary_csv = os.path.join(OUTDIR, "summary_stats.csv")
df_sum = (df_points.groupby(["proteome_id","proteome_name","category","ensemble"], as_index=False)
          .agg(H1_mean=("H1","mean"), H1_std=("H1","std"),
               I1_mean=("I1","mean"), I1_std=("I1","std"),
               Hrate_mean=("Hrate","mean"), Hrate_std=("Hrate","std"),
               Cgz_mean=("Cgz","mean"), Cgz_std=("Cgz","std"),
               Cgz_L_mean=("Cgz_L","mean")))
df_sum.to_csv(summary_csv, index=False)

print("\nSaved:")
print(" -", points_csv)
print(" -", id_csv)
print(" -", summary_csv)

# ----------------------------
# FIGURE 1: H1 vs I1 bootstrap clouds
# ----------------------------
plt.figure(figsize=(10,7))
for pid in df_points["proteome_id"].unique():
    name = proteome_data[pid]["name"]
    for ens in ["Real","IID","Markov-1"]:
        sub = df_points[(df_points["proteome_id"]==pid) & (df_points["ensemble"]==ens)]
        plt.scatter(sub["H1"], sub["I1"], s=8, alpha=0.18, label=f"{name} | {ens}")

plt.axvline(HMAX_BITS, linestyle="--", linewidth=1)
plt.axhline(0.0, linestyle=":", linewidth=1)
plt.xlabel("Composition entropy $H_1$ (bits/site)")
plt.ylabel("Adjacent mutual information $I_1$ (bits)")
plt.title("Bootstrap clouds: Real vs IID vs Markov-1 (20 proteomes)")
handles, labels = plt.gca().get_legend_handles_labels()
seen=set(); h2=[]; l2=[]
for h,l in zip(handles,labels):
    if l not in seen:
        seen.add(l); h2.append(h); l2.append(l)
plt.legend(h2, l2, fontsize=7, ncol=2, frameon=True)
plt.tight_layout()
fig1 = os.path.join(OUTDIR, "Figure1_H1_vs_I1_bootstrap.png")
plt.savefig(fig1, dpi=300, bbox_inches="tight")
plt.show()
print("Saved:", fig1)

# ----------------------------
# FIGURE 2: Id profiles mean±std (one panel per proteome)
# ----------------------------
for pid in df_id_sum["proteome_id"].unique():
    name = proteome_data[pid]["name"]
    plt.figure(figsize=(8,5))
    for ens in ["Real","IID","Markov-1"]:
        sub = df_id_sum[(df_id_sum["proteome_id"]==pid) & (df_id_sum["ensemble"]==ens)].sort_values("d")
        plt.plot(sub["d"], sub["Id_mean"], label=ens)
        plt.fill_between(sub["d"], sub["Id_mean"]-sub["Id_std"], sub["Id_mean"]+sub["Id_std"], alpha=0.15)
    plt.xlabel("Separation $d$ (residues)")
    plt.ylabel("Mutual information $I_d$ (bits)")
    plt.title(f"Information profile $I_d$ (mean ± sd): {name}")
    plt.legend()
    plt.tight_layout()
    fig2 = os.path.join(OUTDIR, f"Figure2_Id_profiles_{pid}.png")
    plt.savefig(fig2, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", fig2)

# ----------------------------
# FIGURE 3: Compressibility gap relative to IID (C_iid - C_real)
# ----------------------------
real = df_points[df_points["ensemble"]=="Real"].copy()
iid  = df_points[df_points["ensemble"]=="IID"].copy()
merged = pd.merge(real, iid, on=["proteome_id","proteome_name","category","rep"], suffixes=("_real","_iid"))
merged["Delta_Cgz"] = merged["Cgz_iid"] - merged["Cgz_real"]

plt.figure(figsize=(10,6))
for pid in merged["proteome_id"].unique():
    name = proteome_data[pid]["name"]
    sub = merged[merged["proteome_id"]==pid]
    plt.scatter(np.arange(len(sub)), sub["Delta_Cgz"], s=8, alpha=0.35, label=name)

plt.axhline(0.0, linestyle=":", linewidth=1)
plt.ylabel(r"$\Delta C_{\mathrm{gzip}} = C_{\mathrm{iid}} - C_{\mathrm{real}}$ (bits/char)")
plt.xlabel("Bootstrap replicate index (stacked by proteome)")
plt.title("Compressibility gap vs IID control (gzip, overhead-corrected)")
plt.legend(fontsize=7, ncol=2, frameon=True)
plt.tight_layout()
fig3 = os.path.join(OUTDIR, "Figure3_Compressibility_gap.png")
plt.savefig(fig3, dpi=300, bbox_inches="tight")
plt.show()
print("Saved:", fig3)

# ----------------------------
# FIGURE 4: Entropy rate comparison (Real only)
# ----------------------------
plt.figure(figsize=(10,6))
for pid in df_points["proteome_id"].unique():
    name = proteome_data[pid]["name"]
    sub = df_points[(df_points["proteome_id"]==pid) & (df_points["ensemble"]=="Real")]
    plt.scatter(sub["H1"], sub["Hrate"], s=8, alpha=0.35, label=name)

plt.axvline(HMAX_BITS, linestyle="--", linewidth=1)
plt.axhline(HMAX_BITS, linestyle="--", linewidth=1)
plt.xlabel("$H_1$ (bits/site)")
plt.ylabel("$H_{rate}^{(1)} = H_1 - I_1$ (bits/site)")
plt.title("Markov-1 entropy rate vs composition entropy (Real sequences) - 20 proteomes")
plt.legend(fontsize=7, ncol=2, frameon=True)
plt.tight_layout()
fig4 = os.path.join(OUTDIR, "Figure4_Entropy_rate.png")
plt.savefig(fig4, dpi=300, bbox_inches="tight")
plt.show()
print("Saved:", fig4)

print("\nAll outputs in:", OUTDIR)
print("Files:", [f for f in os.listdir(OUTDIR) if f.endswith(('.csv','.png'))])

# Optional download (Colab)
try:
    from google.colab import files
    for fn in os.listdir(OUTDIR):
        if fn.endswith(".png") or fn.endswith(".csv"):
            files.download(os.path.join(OUTDIR, fn))
except Exception as e:
    print("Download skipped:", e)

print("\nSummary (head):")
display(df_sum.head(30))

print("\nNote on gzip:")
print(f"  gzip overhead (empty input) = {GZIP_OVERHEAD_BITS} bits")
print("  If Cgz looks unusually high/noisy: increase COMP_MIN_TOTAL_AA and/or COMP_SEQ_CAP.")
```

```python
# ============================
# FIGURE 4 (MAIN): Markov-1 entropy rate vs composition entropy
# Real sequences only
# Summary per proteome: mean ± sd across bootstrap replicates
# Input: bootstrap_points*.csv
# Output: Figure4_Hrate_vs_H1_clean.png
# ============================

from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math


# 2) Real only
real = df[df["ensemble"] == "Real"].copy()

# Si por algún motivo Hrate no existe, lo calculamos
if "Hrate" not in real.columns:
    if ("H1" not in real.columns) or ("I1" not in real.columns):
        raise ValueError("No encuentro columnas Hrate ni (H1, I1).")
    real["Hrate"] = real["H1"] - real["I1"]

# 3) Summary per proteome
sum_pr = (real.groupby(["proteome_id","proteome_name","category"], as_index=False)
              .agg(H1_mean=("H1","mean"), H1_std=("H1","std"),
                   Hrate_mean=("Hrate","mean"), Hrate_std=("Hrate","std")))

# Colors by category (matplotlib default cycle)
categories = list(sum_pr["category"].unique())
cat_to_color = {c: f"C{i%10}" for i,c in enumerate(categories)}

# 4) Plot clean with error bars
plt.figure(figsize=(9,7))

for _, r in sum_pr.iterrows():
    plt.errorbar(
        r["H1_mean"], r["Hrate_mean"],
        xerr=r["H1_std"], yerr=r["Hrate_std"],
        fmt="o", markersize=6, capsize=2,
        alpha=0.9, linewidth=0.8,
        color=cat_to_color[r["category"]]
    )

# Reference lines
xmin = min(sum_pr["H1_mean"].min(), sum_pr["Hrate_mean"].min())
xmax = max(sum_pr["H1_mean"].max(), sum_pr["Hrate_mean"].max())
pad = 0.02
lo, hi = xmin - pad, xmax + pad

# y = x diagonal
plt.plot([lo, hi], [lo, hi], linestyle="--", linewidth=1)

# Theoretical maximum log2(20)
Hmax = math.log2(20)
plt.axvline(Hmax, linestyle=":", linewidth=1)
plt.axhline(Hmax, linestyle=":", linewidth=1)

plt.xlim(lo, hi)
plt.ylim(lo, hi)

plt.xlabel("Composition entropy $H_1$ (bits/site)")
plt.ylabel(r"First-order entropy rate $H^{(1)}_{\mathrm{rate}} = H_1 - I_1$ (bits/site)")
plt.title(r"Local ordering reduces per-residue uncertainty: $H^{(1)}_{\mathrm{rate}}$ vs $H_1$ (Real proteomes)")

# Legends: categories (colors)
handles_cat = [
    plt.Line2D([0],[0], marker='o', linestyle='', color=cat_to_color[c], label=c)
    for c in categories
]
plt.legend(handles=handles_cat, title="Category", loc="lower left", fontsize=9, frameon=True)

plt.tight_layout()
out_png = "Figure4_Hrate_vs_H1_clean.png"
plt.savefig(out_png, dpi=300, bbox_inches="tight")
plt.show()

print("Saved:", out_png)
files.download(out_png)
```

```python
# ============================================================
# COLAB NOTEBOOK CELL (RUN ALL)
# Upload CSV(s) from your PC, then generate 20 Id-profile figures
# ============================================================

import os, re, shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from google.colab import files

# ----------------------------
# 1) Upload CSV from your PC
# ----------------------------
print("Upload your CSV from your PC (Id_profile_mean_std (1).csv).")
uploaded = files.upload()  # opens file picker

# Find the uploaded Id_profile file automatically
def find_uploaded(pattern):
    for fn in uploaded.keys():
        if pattern.lower() in fn.lower():
            return "/content/" + fn
    # fallback: search in /content
    for fn in os.listdir("/content"):
        if pattern.lower() in fn.lower():
            return "/content/" + fn
    return None

CSV_PATH = find_uploaded("Id_profile_mean_std")
if CSV_PATH is None:
    raise FileNotFoundError(
        "I couldn't find a file with 'Id_profile_mean_std' in its name. "
        "Please upload Id_profile_mean_std (1).csv."
    )

print("Using:", CSV_PATH)

# ----------------------------
# 2) Load and validate
# ----------------------------
df = pd.read_csv(CSV_PATH)

required = {"proteome_id","proteome_name","category","ensemble","d","Id_mean","Id_std"}
missing = required - set(df.columns)
if missing:
    raise ValueError(f"Missing required columns: {missing}")

df["d"] = pd.to_numeric(df["d"], errors="coerce").astype(int)
df["Id_mean"] = pd.to_numeric(df["Id_mean"], errors="coerce")
df["Id_std"]  = pd.to_numeric(df["Id_std"], errors="coerce")

# Keep only sensible ds
df = df[df["d"].between(1, 100)].copy()

print("Loaded:", df.shape)
display(df.head())

# ----------------------------
# 3) Output folders
# ----------------------------
OUTDIR = "/content/Id_profiles_20_outputs"
PNGDIR = os.path.join(OUTDIR, "PNGs")
os.makedirs(PNGDIR, exist_ok=True)

PDF_PATH = os.path.join(OUTDIR, "All_20_IdProfiles.pdf")

# ----------------------------
# 4) Plot configuration
# ----------------------------
ENSEMBLE_ORDER = ["Real", "IID", "Markov-1"]
COLOR = {"Real":"C0", "IID":"C1", "Markov-1":"C2"}  # default matplotlib colors

def safe_filename(s: str, maxlen: int = 80) -> str:
    s = re.sub(r"[^A-Za-z0-9._-]+", "_", str(s)).strip("_")
    return s[:maxlen] if len(s) > maxlen else s

def plot_id_profile_one_proteome(sub: pd.DataFrame, title: str):
    """
    Single figure for one proteome:
    lines + mean±std shading for Real / IID / Markov-1
    """
    sub = sub.copy()
    ds = np.sort(sub["d"].unique())

    fig, ax = plt.subplots(figsize=(6.6, 4.2))

    for ens in ENSEMBLE_ORDER:
        s2 = sub[sub["ensemble"] == ens].sort_values("d")
        if s2.empty:
            continue

        x = s2["d"].values
        y = s2["Id_mean"].values
        e = s2["Id_std"].values

        ax.plot(x, y, label=ens, color=COLOR.get(ens, None), linewidth=2)
        ax.fill_between(x, y - e, y + e, color=COLOR.get(ens, None), alpha=0.18)

    ax.set_title(title)
    ax.set_xlabel("Separation d (residues)")
    ax.set_ylabel("Mutual information $I_d$ (bits)")
    ax.set_xlim(ds.min(), ds.max())
    ax.grid(True, alpha=0.25)
    ax.legend(frameon=True, fontsize=9)
    fig.tight_layout()
    return fig

# ----------------------------
# 5) Generate 20 PNGs + 1 PDF
# ----------------------------
proteomes = (df[["proteome_id","proteome_name"]]
             .drop_duplicates()
             .sort_values("proteome_name"))

print("Proteomes found:", len(proteomes))

with PdfPages(PDF_PATH) as pdf:
    for _, row in proteomes.iterrows():
        pid = row["proteome_id"]
        pname = row["proteome_name"]

        sub = df[df["proteome_id"] == pid]
        if sub.empty:
            continue

        fig = plot_id_profile_one_proteome(sub, title=str(pname))

        png_name = f"IdProfile_{safe_filename(pname)}_{pid}.png"
        png_path = os.path.join(PNGDIR, png_name)
        fig.savefig(png_path, dpi=300, bbox_inches="tight")

        pdf.savefig(fig)
        plt.close(fig)

print("DONE.")
print("PNG folder:", PNGDIR)
print("PDF:", PDF_PATH)

# ----------------------------
# 6) Zip outputs and download
# ----------------------------
ZIP_PATH = "/content/Id_profiles_20_outputs.zip"
shutil.make_archive(ZIP_PATH.replace(".zip",""), "zip", OUTDIR)

print("ZIP ready:", ZIP_PATH)
files.download(ZIP_PATH)
```
